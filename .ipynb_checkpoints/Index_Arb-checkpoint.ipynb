```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("aapl_amz_koyfin.csv")

print(df.columns.tolist())
df['Date']= pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df['r_aapl'] = df['XLK Close']/ df['AAPL Close']
df['r_amzn'] = df['XLK Close']/ df['AMZN Close'].astype(float)

fig, (ax0,ax1) = plt.subplots(2,1, figsize=(100,50))
df['r_aapl'].plot(ax=ax0)
df['r_amzn'].plot(ax=ax1)
plt.show()
plt.close()
```

```python
## cointegrated? copula? cointegrated with a drift?
## set drift as it's moving average, will that make spread/ residual stationary?

df['r_aapl'] = df['r_aapl'] - df['r_aapl'].rolling(30).mean()
df['r_amzn'] = df['r_amzn'] - df['r_amzn'].rolling(30).mean()
fig, (ax0,ax1) = plt.subplots(2,1, figsize=(100,50))
df['r_aapl'].plot(ax=ax0)
df['r_amzn'].plot(ax=ax1)
plt.show()
plt.close()

```

```python
## test for stationarity, ADF
from statsmodels.tsa.stattools import adfuller

print(adfuller(df['r_aapl'].dropna()))
print(adfuller(df['r_amzn'].dropna()))
```

```python
## define moving windows/ rolling mu_t and std dev_t 
## long if r cross lower bound:= mu_t + 1.6945*std dev_t
## short if r cross upper bound:= mu_t - 1.6945*std dev_t
## reson for moving average again as the ratio still has mean, but reverting quicker than before

df['UB_aapl'] = df['r_aapl'].rolling(30).mean() + 1.645*df['r_aapl'].rolling(30).std()
df['LB_aapl'] = df['r_aapl'].rolling(30).mean() - 1.645*df['r_aapl'].rolling(30).std()

df['UB_amzn'] = df['r_amzn'].rolling(30).mean() + 1.645*df['r_amzn'].rolling(30).std()
df['LB_amzn'] = df['r_amzn'].rolling(30).mean() - 1.645*df['r_amzn'].rolling(30).std()


# df['UB_aapl'] = 1.645*df['r_aapl'].rolling(30).std()
# df['LB_aapl'] = - 1.645*df['r_aapl'].rolling(30).std()

# df['UB_amzn'] =  + 1.645*df['r_amzn'].rolling(30).std()
# df['LB_amzn'] = - 1.645*df['r_amzn'].rolling(30).std()



```

```python
fig, (ax0,ax1) = plt.subplots(2,1, figsize=(100,50))
df['r_aapl'].plot(ax=ax0)
df['UB_aapl'].plot(ax=ax0,c='r')
df['LB_aapl'].plot(ax=ax0, c='g')


df['r_amzn'].plot(ax=ax1)
df['UB_amzn'].plot(ax=ax1,c='r')
df['LB_amzn'].plot(ax=ax1, c='g')
plt.show()
plt.close()
```

```python
for ticker in ['aapl','amzn']:
    df.loc[df[f"r_{ticker}"]<=df[f"LB_{ticker}"], 'signal_aapl'] = 1
    df.loc[df[f"r_{ticker}"]>=df[f"UB_{ticker}"], 'signal_amzn'] = -1
    
```
